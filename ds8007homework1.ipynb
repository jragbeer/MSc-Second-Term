```python
import pandas as pd
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
age_of_students = np.array([3, 11, 12, 19, 22, 23, 24, 25, 27, 29, 15, 6, 27, 17, 29])
plt.boxplot(age_of_students, labels = ['Age of Students'], meanline = True, showmeans=True)

print("Question: What is the average age of the students in the class?")
print("We can see from this box plot a lot of characteristics of the class. We can immediately see the oldest, youngest, the quartiles and the mean (green dotted line).")
```

```python
subjectdict = {'French': 30,'English': 20,'Maths':26,'Geography': 38,'Science':34}
plt.pie(list(subjectdict.values())[::-1],labels = list(subjectdict.keys())[::-1], colors = ['red', 'tomato', 'firebrick', 'darkseagreen', 'lightgreen'])

print("Question: Do more students like STEM subjects or the languages?")
print('We can see from this pie chart that there is a larger amount of students that prefer a STEM subject (red) versus a language (green).')
```

```python
schooldict = {"School": 6, "Sleep": 8, "Playing": 2, "Study": 4, "T. V.": 1, "Others": 3}

plt.bar(list(schooldict.keys()), list(schooldict.values()))

print("Question: Do children still enjoy T.V.?")
print("We can see from this bar chart that T.V. isn't that popular among the children. It is behind many other activities.")
```

```python
A=[0.22, -0.87, -2.39, -1.79, 0.37, -1.54, 1.28, -0.31, -0.74, 1.72, 0.38, -0.17, -1.62, -1.10, 0.30, 0.15, 2.30, 0.19, -0.50, -0.09]
B=[-5.13, -2.15, -3.43, -3.83, 0.50, -3.25, 4.32, 0.63, 5.18, -0.43, 7.11, 4.87, -3.10, -5.81, 3.76, 6.31, 2.58, 0.07, 5.76, 3.50]

plt.plot(range(len(A)), A, label = 'Group A', c = 'green')
plt.plot(range(len(B)), B, label = 'Group B', c = 'red')
plt.plot(range(len(A)), [np.mean(A) for x in A], alpha = 0.23, c = 'green', label = 'Group A mean')
plt.plot(range(len(B)), [np.mean(B) for x in B], alpha = 0.23, c = 'red', label = 'Group B mean')
plt.legend()

print("Question: Which group is performing better (high values are better)?")
print('We can see from this set of line charts that Group B is performing better. Group B has a better average, as well as the 5 out of the 5 highest scores.')
```
